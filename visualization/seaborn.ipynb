```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df
```

Survival and Non - Survival count by Gender

```python
sns.countplot(df , x="alive" , hue="sex")
plt.title("Titanic survivors")
plt.show()
```

```python
sns.countplot(df , y="class" , palette="viridis")
plt.title("Titanic class")
plt.show()
```

```python
sns.barplot(df , x="class", y="survived",palette="coolwarm")
plt.title("Survivors by class")
plt.show()
```

```python
sns.histplot(df , x="age", bins=30, kde=True, color="red")
plt.title("Age distribution")
plt.show()
```

```python
sns.violinplot(df, x='class',y ='age', hue='sex', palette='coolwarm')
plt.title("Age distribution by class")
plt.show()
```

```python
df = pd.read_csv("country_wise_latest.csv")
df
```

Top 10 countries with higher number of cases

```python
top_10 = df.nlargest(10, 'Confirmed')
sns.barplot(x = "Confirmed" , y="Country/Region" , data=top_10, palette="coolwarm")
plt.title("Top 10 countries with most confirmed cases")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Confirmed last week', y='1 week change', data=df, marker='o')
plt.title('Weekly Case Growth')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='Confirmed', y='Active', data=df, hue='WHO Region', palette='viridis')
plt.title('Active Case vs Confirmed case')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['Deaths / 100 Cases'], bins = 30, kde=True, color='red')
plt.title("Distribution of Death rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='WHO Region', y='New cases', data=df, palette='pastel')
plt.xticks(rotation=45)
plt.title('New Cases by WHO Region')
plt.show()
```

```python
plt.figure(figsize=(8,8))
df.groupby('WHO Region')['Confirmed'].sum().plot.pie(autopct='%1.1f%%', cmap='coolwarm', startangle=140)
plt.ylabel('')
plt.title('Confirmed Cases Distribution by WHO Region')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.relplot(data=df, x="Deaths", y="Active",hue="WHO Region",height=6)
plt.title("CASES")
plt.show()
```

```python

sns.pairplot(df[['Confirmed', 'Deaths', 'Recovered', 'Active']], diag_kind='kde', plot_kws={'alpha':0.5})
plt.show()
```

```python
sns.violinplot(data=df, x="WHO Region", y="Deaths / 100 Recovered")
plt.title("Deaths per 100 Recovered by WHO Region")
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.swarmplot(data=df, x="WHO Region", y="Recovered / 100 Cases")
plt.title("Recovered per 100 Cases by WHO Region")
plt.xticks(rotation=45)
plt.show()
```
